## What can you do with NLP?

Natural Language Processing (NLP) projects often require the implementation of a range of fundamental tasks. These tasks, due to their repetitive and foundational role across various NLP applications, have been widely studied and expertise in them can significantly enhance your ability to build effective NLP solutions. Below, we will get into each task individually, offering a detailed overview of its function, importance, and application.

### 1. Language Modeling

Language modeling revolves around predicting the next word in a sentence based on preceding words. This task is foundational for many NLP applications, such as speech recognition and machine translation. By learning the probability of word sequences, language models can generate coherent and contextually relevant text.

### 2. Text Classification

Text classification involves categorizing text into predefined categories based on its content. This is one of the most common tasks in NLP and has numerous applications, such as spam detection, sentiment analysis, and topic classification. By training models to recognize patterns in text, we can automate the process of sorting and labeling large volumes of text data.

### 3. Information Extraction

Information extraction focuses on extracting relevant information from text. For example, it can identify dates and events from emails or extract names and locations from news articles. This task is essential for transforming unstructured text into structured data that can be easily analyzed and queried.

### 4. Information Retrieval

Information retrieval is about finding relevant documents based on a user's query. This task is the backbone of search engines like Google, where the goal is to deliver the most relevant results from a vast collection of documents. Effective information retrieval requires understanding both the user's query and the content of the documents to provide accurate matches.

### 5. Conversational Agents

Conversational agents, or chatbots, are designed to interact with users in natural language. These systems can range from simple rule-based bots to advanced AI-driven assistants like ChatGPT and Siri. The goal is to create a seamless and intuitive user experience where the system can understand and respond to user queries effectively.

### 6. Text Summarization

Text summarization aims to create concise summaries of longer documents while preserving the central meaning. This is particularly useful for quickly understanding the gist of lengthy reports, articles, or research papers without reading the entire text. Summarization can be either extractive (selecting key sentences) or abstractive (generating new sentences).

### 7. Question Answering

Question answering systems are designed to automatically answer questions posed in natural language. These systems can range from simple FAQs to complex AI-driven models that understand context and provide detailed answers. Effective question answering requires a deep understanding of language and the ability to retrieve and integrate information from various sources.

### 8. Machine Translation

Machine translation involves translating text from one language to another. This task is essential for breaking down language barriers and enabling global communication. Advanced models like Google Translate use sophisticated algorithms to provide accurate and contextually appropriate translations.

### 9. Topic Modeling

Topic modeling is about uncovering the hidden thematic structure within a large set of documents. This technique is widely used in text mining to discover the main topics discussed in a corpus. Applications range from analyzing literature and academic papers to understanding customer feedback and social media trends.

## Corpus in Natural Language Processing (NLP)

In Natural Language Processing (NLP), a **corpus** (plural: _corpora_) is a fundamental concept. It refers to a carefully compiled collection of text data, sourced from various channels, and written in human language – what we call *natural language*. This data serves as the foundation for training and evaluating NLP models, enabling them to understand and process human language effectively. 

###  Size and Scope of Corpora

Corpora exhibit vast diversity in their size, ranging from compact collections to massive repositories comprising billions of words. The choice of corpus size often depends on the specific NLP task and the computational resources available. For instance, a simple sentiment analysis task might suffice with a smaller, domain-specific corpus, while training a sophisticated language model like those used in machine translation would demand a massive corpus encompassing diverse writing styles and domains. 

To illustrate, the widely used [The Pile](https://huggingface.co/datasets/EleutherAI/pile) corpus contains a staggering 825 GB of English text data, demonstrating the scale these datasets can reach.  It's not uncommon for researchers and practitioners to work with corpora containing hundreds of gigabytes or even terabytes of text data, especially when dealing with fields rich in textual information, such as law or scientific research. 

### Types of Corpora: Annotated vs. Unannotated

Corpora can be broadly classified into two main types based on the presence or absence of labels:

1. **Annotated Corpora:**  These corpora go beyond raw text by incorporating **annotations** – labels or tags added to the text to indicate specific linguistic phenomena or semantic information. 
    - **Example:** In a sentiment analysis task, an annotated corpus might have each sentence labeled as positive, negative, or neutral, providing the ground truth for training a model to recognize sentiment. 

2. **Unannotated Corpora:** As the name suggests, these corpora consist solely of raw text without any added labels. 
    - **Example:** A large collection of news articles without any topic labels would constitute an unannotated corpus.

**The Role of Annotation in Training:**

- Annotated corpora are essential for **supervised learning** tasks, where the model learns to make predictions by generalizing from labeled examples.
- Unannotated corpora, while lacking explicit labels, are valuable for **unsupervised learning**, where models uncover patterns and structures in data without explicit guidance.

> **Note**: As you can imagine, annotating data is expensive. Because of this, most deep learning approaches today uses unannotated and annotated data to train models. For example, a model trained to classify news articles into various categories might be trained on a large unannotated corpus of news articles and then fine-tuned on a smaller annotated corpus of news articles. We'll see more about this in the upcoming lessons, when we'll discuss _transfer learning_ but for now, it's important to understand the difference between annotated and unannotated corpora.


### Structure of a Corpus: From Documents to the Smallest Units

A corpus is not simply a massive, unstructured text file.  It is typically organized hierarchically, breaking down the data into progressively smaller units of meaning, which are crucial for NLP tasks.  Understanding this structure is key to working with and analyzing textual data effectively.  Here's a breakdown:

1. **Documents:** The highest level of organization consists of individual documents. Each document represents a self-contained unit of text, conveying a coherent set of ideas or information. For example, a news article, a research paper, or a social media post could all be considered individual documents within a corpus. 

2. **Paragraphs:**  Documents are further divided into paragraphs, each addressing a specific aspect or point within the broader theme of the document.

3. **Sentences:** Paragraphs, in turn, break down into sentences – the fundamental units of language that express complete thoughts.

4. **Words and Punctuation:** Sentences comprise individual words, carrying semantic meaning, and punctuation marks, which provide grammatical context and structure to the text.

5. **Syllables, Phonemes, Morphemes, and Characters:**  Delving deeper, we encounter the building blocks of words themselves.
   -  **Syllables:** Units of pronunciation (e.g., "nat-u-ral").
   -  **Phonemes:** Individual sound units (e.g., the "n" sound in "natural").
   -  **Morphemes:**  The smallest units of meaning (e.g., "un-" to indicate negation).
   -  **Characters:**  The individual letters or symbols used in writing (e.g., "a", "b", "c"). 

This hierarchical structure allows NLP models to analyze text at various levels of granularity, capturing both the overall meaning of a document and the nuances conveyed by individual words and their components.

### Domain-Specific Corpora: Focusing on Specialized Language

While general-purpose corpora are vast and encompass diverse writing styles, they might not be ideal for tasks that require an understanding of specialized language.  This is where **domain-specific corpora** come into play. These are corpora specifically designed to capture the language, terminology, and conventions of a particular field or industry. 

**Examples of Domain-Specific Corpora:**

* **Medical Corpus:** Might include clinical notes, research articles, medical textbooks, and patient records, all using specialized medical vocabulary.
* **Legal Corpus:**  Could consist of court transcripts, legal contracts, law journals, and legislation, reflecting the formal and precise language used in the legal domain.
* **Financial Corpus:**  Might contain financial news articles, company reports, stock market data, and economic analyses, using domain-specific financial jargon and reporting conventions.

**Advantages of Domain-Specific Corpora:**

* **Improved Accuracy:**  By training on data closely resembling the target domain, NLP models achieve higher accuracy in tasks like text classification, information extraction, and machine translation within that specific field.
* **Deeper Insights:**   Analyzing a corpus dedicated to a particular field allows researchers to uncover hidden patterns, trends, and insights specific to that domain, which might not be apparent in a general-purpose corpus.

**Challenges:**

* **Data Collection:** Building a high-quality domain-specific corpus requires careful selection and curation of relevant documents from credible sources, which can be a time-consuming and resource-intensive process.
* **Annotation:**   Annotating specialized corpora often demands domain experts who can accurately label the text with relevant tags and categories.


### Building a Domain-Specific Corpus

Constructing a valuable domain-specific corpus is an iterative process requiring careful planning and execution. Here are the key steps involved:

1. **Define the Domain:**  Begin by precisely outlining the specific field or industry the corpus will focus on. This clarity is essential to guide data collection and ensure relevance.

2. **Data Collection:**  Gather a diverse range of relevant documents from reputable sources within the chosen domain. This could involve accessing public databases, scraping websites, obtaining permissions for copyrighted materials, or collaborating with institutions.

3. **Data Cleaning:** This critical step involves removing irrelevant information, such as advertisements, duplicate content, or boilerplate text, ensuring the corpus contains only high-quality, relevant data.

4. **Preprocessing:**   Prepare the text data for analysis by:
   - Converting to lowercase:  Ensures uniformity and prevents the model from treating words with different capitalization as distinct. 
   -  **Stemming/Lemmatization:** Reducing words to their root form (e.g., "running," "ran," and "runs" all become "run") for standardized analysis.
   - Removing stop words:  Eliminating common words like "the," "a," and "is," which carry less meaning in the context of the target task.
   -  **Tokenization:** Breaking down text into individual words or units of meaning, facilitating further processing.

5. **Categorization/Classification:**   Organize the corpus by categorizing documents based on relevant criteria within the chosen domain. This helps in targeted analysis and training of models for specific subtasks.

6. **Validation:** Rigorously evaluate the corpus's quality, ensuring the documents are relevant, the cleaning process was effective, and the categorization aligns with the intended use. Validation ensures the corpus is a reliable and valuable resource for NLP tasks.


Bear in mind that creating a domain-specific corpus is an iterative process, where you might need to revisit these steps until optimal results are achieved.

Approaches typically use BIO notation, which differentiates the beginning (B) and the inside (I) of entities. O is used for non-entity tokens. Let us look at an example with [LeNER-Br](https://huggingface.co/datasets/lener_br), which is a dataset for NER in Brazilian legal documents.

````python
from datasets import load_dataset

ds_lener = load_dataset("lener_br", trust_remote_code=True)
````

````python
ds_lener
````

````python
ds_lener['train'][0]
````

````python
# Create a dictionary to map integer NER tag values to their corresponding string names
# This is useful for interpreting the NER tags in a human-readable format

# 'ds_lener' is assumed to be a dataset with a 'train' split containing 'features' which include 'ner_tags'
# 'str2int' converts string NER tags to their integer representations
# 'names' provides the list of string NER tag names

mapping_key_lener = {
    ds_lener['train'].features['ner_tags'].feature.str2int(k): k 
    for k in ds_lener['train'].features['ner_tags'].feature.names
}

# Display the mapping dictionary to verify its contents
mapping_key_lener
````

````python
# Define a function to display tokens and their corresponding NER tags for a given sentence index
# Parameters:
# - idx: Index of the sentence to display
# - ds: The dataset containing the sentences and their NER tags
# - mapping_key: Dictionary mapping integer NER tags to their string names
def show_tagged_sentence(idx, ds, mapping_key):
    # Iterate over the tokens and their corresponding NER tags in the selected sentence
    for token, tag in zip(ds['train'][idx]['tokens'], ds['train'][idx]['ner_tags']):
        # Only print the token and its tag if the tag is not 0 (assuming 0 means 'no tag')
        if tag != 0:
            print(f'{token}\t\t{mapping_key[tag]}')

# Call the function to display the tokens and NER tags for the first sentence in the dataset
show_tagged_sentence(0, ds_lener, mapping_key_lener)
````

````python
show_tagged_sentence(1, ds_lener, mapping_key_lener)
````

### Building Your Own Corpora

Building a corpus, the large and structured set of texts that are used in natural language processing and machine learning, can be as straightforward or as intricate as you need it to be. Let's understand step-by-step how you can create a custom corpus.

#### Step 1: Setting Up Your Corpus

At its most basic, a corpus could simply be a directory on your computer system, filled with plain text files. These would be the documents that your model will learn from.

The range of options is virtually limitless!

#### Step 3: Case Study - Extracting Data from a PDF

In this example, let's focus on creating a corpus from a drug leaflet for Metamizole (Dipyrone) which is available online in a PDF format.

To do this, we'll leverage Python's `requests` library to download the HTML pages and the [`PyPDF2`](https://pythonhosted.org/PyPDF2/) package to extract the text content from these PDF files.

By following this process, you should be able to build a varied and rich corpus, tailored to suit your project's requirements. Don't hesitate to experiment with different types of files and sources. The broader and more diverse your corpus, the better your model will perform on unseen data!

````python
import requests
import PyPDF2
from pathlib import Path

# Download the PDF file from the URL
url = 'https://www.ache.com.br/wp-content/uploads/application/pdf/bula-paciente-dipirona-monoidratada-gotas.pdf'
response = requests.get(url, verify=False) # we set verify to False to avoid SSL certificate errors

# Create the tmp path if it doesn't exist
Path('outputs/tmp').mkdir(parents=True, exist_ok=True)

# Save the file locally
with open('outputs/tmp/bula.pdf', 'wb') as f:
    f.write(response.content)

# Create a PDF reader object from the downloaded content
pdf_reader = PyPDF2.PdfReader('outputs/tmp/bula.pdf')

# Get the number of pages in the PDF document
num_pages = len(pdf_reader.pages)

# Loop through each page and extract the text
for page_num in range(num_pages):
    # Get the page object
    page = pdf_reader.pages[page_num]

    # Extract the text from the page
    text = page.extract_text()

    # Print the text
    print(text)
````

You can also scrape the data from web and parse with BeautifulSoup. Here is an example of scraping the data from the web and parsing it with BeautifulSoup.

````python
import requests
from bs4 import BeautifulSoup

# Define the URL of the Wikipedia article
url = 'https://en.wikipedia.org/wiki/Metamizole'

# Send a GET request to the URL and get the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the main text content of the article
article_content = soup.find('div', {'class': 'mw-parser-output'})

# Remove the table of contents and other non-text elements
for element in article_content(['table', 'sup', 'span', 'style']):
    element.decompose()

# Extract the text from the article content
text = article_content.get_text()

# Print the text
print(text.strip())
````

### Step 3: Preparing the Data for Annotation

Proper data preparation is crucial for a smooth annotation process. Label Studio requires data to be in a specific JSON format:

```json
[
    {
        "id": 1,
        "data": {
            "text": "Text content goes here",
            "meta_info": {
                "meta1": "value1",
                "meta2": "value2"
            }
        }
    }
]
```

Key components of this format include:

- **id:** A unique identifier for each data point.
- **data:** Contains the actual content to be annotated.
  - **text:** The main text or content for annotation.
  - **meta_info:** Additional metadata that may be relevant to the annotation task.

To start, we'll use the first 100 lines from our dataset `ds_b2w` and convert it into the required format. The metadata (`meta_info`) can include additional information that could be useful for the labeling task.
